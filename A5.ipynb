```python
#import necessary libraries
import numpy as np
import glob 
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib import rcParams
```

```python
#Create a function called visualization 
def visualize(filename):
    
    "Load dataset and assign header as none as there is no header in our dataset"
    d = pd.read_csv(filepath_or_buffer = filename, header=None)
    "calculate maximum of coloumn"
    maximum = d.apply(np.max)
    
    "calculate minimum of coloumn"
    minimum = d.apply(np.min)
    
    "calculate average of coloumn"
    average = d.iloc[:,:].apply(np.mean)
    
    "Set figure size"
    rcParams['figure.figsize']=10,5
    
    "Plot line graph with appropriate labels"
    plt.plot(maximum, label='Maximum')
    plt.plot(average, label='Average')
    plt.plot(minimum, label='Minimum')
    plt.legend(loc=7)
    plt.grid()
    plt.title(filename)
    plt.xlabel("Feature#")
    plt.ylabel("Values")
    plt.show()
```

```python
#Create issue identifying function 
def Identify_Issue(filename):
    d = pd.read_csv(filepath_or_buffer = filename, header=None) #read data
    minimum = d.apply(np.min) #Calculate min of coloumn
    if d.iloc[:,0].max() == 0 and d.iloc[:,20].max() == 20: #Check the condition and both should fulfill 
        print('I doubt the integrity')
    else:
        if minimum.sum() == 0: #Check if sum of all min value is 0
             print('Why is this happening?')
        else:
            print('Dataset looks good.')
```

```python
# Use glob function to call file starting with data 
data= sorted(glob.glob('data*.csv'))
print(data)
```

```python
#Run for loop with a line space for 4 data files
for datas in data[:4]:
    print("\n")
    print(datas)
    visualize(datas)
    Identify_Issue(datas)
```
